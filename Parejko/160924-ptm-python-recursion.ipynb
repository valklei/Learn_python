```python
def compute_sum(n):
    if n == 0:
        return n
    else:
        return n + compute_sum(n - 1) # 5 + 4 + 3 + 2 + 1 + 0


print(compute_sum(5))
```

```python
def compute_sum_iterative(n):
    sum_ = 1
    
    for i in range(1, n + 1):
        sum_ += i
        
    return sum_

print(compute_sum_iterative(5))
```

```python
def add_item_to_list(n, in_list=[]):
    if n == 0:
        return in_list
    in_list.append(n)
    return add_item_to_list(n - 1, in_list) # хвостовая


result = add_item_to_list(3) # [3, 2, 1]
print(f"List_1: {result}")

result_2 = add_item_to_list(2) # [3, 2, 1, 2, 1]
print(f"List_2: {result_2}")
```

```python
def add_item_to_list(n=7, in_list=None):
    if in_list is None:
        in_list = []
    
    if n == 0:
        return in_list
    in_list.append(n)
    return add_item_to_list(n - 1, in_list) # хвостовая


result = add_item_to_list(3) # [3, 2, 1]
print(f"List_1: {result}")

result_2 = add_item_to_list(2) # [2, 1]
print(f"List_2: {result_2}")
```

```python
# 0, 1, 1, 2, 3, 5

def fibonacci_recursive(n): # O(2^n)
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)


def fibonacci_iterative(n): # O(n)
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
```

```python
import time


target_val = 30

start_time = time.process_time()
result_1 = fibonacci_recursive(target_val)
recursive_time = time.process_time() - start_time


start_time = time.process_time()
result_2 = fibonacci_iterative(target_val)
iterative_time = time.process_time() - start_time

print(f"Время выполнения recursion равно {recursive_time:.6f} секунд")
print(f"Время выполнения iterative равно {iterative_time:.6f} секунд")
```

```python
def deep_copy(obj):
    new_list = []
    
    for item in obj:
        if type(item) == list:
            new_list.append(deep_copy(item))
        else:
            new_list.append(item)
    
    return new_list


new_list_2 = deep_copy(test_list)

new_list_2[1][0] = "WWW"
print(new_list_2)

print(test_list)
```
