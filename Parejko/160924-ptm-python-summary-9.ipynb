```python
def calculate_sum_of_n(n, acc=0):
    if n == 0: # базовый случай
        return acc
    else: # рекурсивный случай
        return calculate_sum_of_n(n - 1, acc + n) # 5 + 4 + 3 + 2 + 1
    
print(calculate_sum_of_n(5))
```

```python
def modify_list(numbers):
    """
    Modifies the list: multiplies odd numbers by 2, and divides even numbers by 2.
    """
    for i in range(len(numbers)):
        if numbers[i] % 2 == 0:
            numbers[i] //= 2 # Делим четное число на 2
        else:
            numbers[i] *= 2 # Умножаем нечетное число на 2

user_input = input("Enter numbers separated by space: ") # -> [-1, '4', 3, 6, '7', 'qwerty', 'hello', 14]

if not all(item.isdigit() or (item.startswith('-') and item[1:].isdigit()) for item in user_input.split()):
    print("Incorrect data: please enter only integers.")
else:
    numbers = list(map(int, user_input.split()))
    modify_list(numbers)
    print("Changed list of numbers:", numbers)
# Списки в Python являются изменяемыми объектами (mutable objects). Это означает, что если передать список в функцию, то изменения, сделанные внутри функции, будут сохраняться в оригинальном списке.
```

```python
def modify_list(numbers): # [2, 3, 5, 6, 3, 8, 9]
    return list(map(lambda x: x * 2 if x % 2 != 0 else x // 2, numbers))


def convert_to_int(x):
    if x.isdigit():
        return int(x)

user_input = input("Введите числа через пробел: ")
numbers = list(map(convert_to_int, user_input.split()))

modified_numbers = modify_list(numbers)

print("Введены следующие числа: ", user_input)
print("Измененный список:", modified_numbers)

```

```python
matrix = []

in_list = input('Enter data [with spaces]:')
cut_list = in_list.split()


for i in cut_list: 
    if i.isdigit(): 
        matrix.append(int(i))
        

print(f"Исходный список чисел: {cut_list}") 
print(f"Корр список чисел: { matrix}") 


def calculate_sum(*args): # -> tuple -> (2, 3, 4, 5, 6)
    return sum(args)

print(f"Сумма введенных чисел: {calculate_sum(*matrix)}")
```

```python
def factor(n):
    if n == 0 or n == 1:
        return 1
    else: 
        return n * factor(n - 1)
    
print(factor(5)) # 5 * 4 * 3 * 2 * 1
```
