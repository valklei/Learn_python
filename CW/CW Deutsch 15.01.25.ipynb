```python
# Erstellen Sie eine Klasse Laptop mit den Feldern: Marke und Farbe. 
# Erstellen Sie 5 Objekte dieser Klasse. 
# Schreiben Sie eine Funktion, die eine Liste von Laptops und eine Farbe als 
# Eingabe erhält und eine Liste von Laptops dieser Farbe zurückgibt.

class Laptop():
    def __init__(self, marke, farbe):
        self.marke = marke
        self.farbe = farbe
        
    def __repr__(self):
        return f"Laptop(Brand: {self.marke}, Color: {self.farbe})"    
   
def farbe_filter(farbe, laptop_list):
    filtered_laptop = []
    for elem in laptop_list:
        if elem.farbe == farbe:
            filtered_laptop.append(elem)
    return filtered_laptop
            
laptop1 = Laptop("Asus", "Grun")
laptop2 = Laptop("Dell", "Schwarz")
laptop3 = Laptop("HP", "Rot")
laptop4 = Laptop("Asus", "Rot")
laptop5 = Laptop("Mac", "Weiß")
laptop6 = Laptop("Dell", "Gelb")

laptop_list = [laptop1, laptop2, laptop3, laptop4, laptop5, laptop6]

result = farbe_filter("Rot", laptop_list)
print(result)
        
```

```python
Erstellen Sie eine Klasse Student mit den Feldern: 
Name und Alter. Erstellen Sie 5 Objekte dieser Klasse. 
Schreiben Sie eine Funktion, die aus einer Liste von Studenten 
nur diejenigen zurückgibt, die älter als 18 Jahre sind

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"Student({self.name}, {self.age})"
    def __repr__(self):
        return f"Student({self.name}, {self.age})"

def student_filter(students, age):
    return list(filter(lambda student: student.age > age, students))

students = [
    Student("Student1", 16),
    Student("Student2", 17),
    Student("Student3", 18),
    Student("Student4", 19),
    Student("Student5", 20)
]

age = 17
print(student_filter(students, age))
print(repr(student_filter(students, age)))
print('-'*30)
for item in student_filter(students, age):
    print(f"Student({item.name}, {item.age})")
```

```python
from datetime import datetime

class Animal:
    def __init__(self, name, geburtsdatum):
        self.name = name
        self.geburtsdatum = geburtsdatum
        self.animals_age = self.calc_animal_age()

    def calc_animal_age(self):
        current_date = datetime.now()
        age_in_human_years = (current_date - self.geburtsdatum).days // 365
        age_in_animal_years = age_in_human_years * 7
        return age_in_animal_years

class Pet:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Pet({self.name}, {self.age})"

def convert_pets(animals):
    pets = []
    for animal in animals:
        if animal.animals_age > 2:
            pet = Pet(animal.name, animal.animals_age)
            pets.append(pet)
    return pets
```
